```python
key = pd.read_csv('data.csv')
key.head()
```

```python
import datetime

def key_date(data_str):
    norm_data_str = data_str.split(';')
    return datetime.datetime.strptime(norm_data_str[0], '%d.%m.%Y')

def key_value(data_str):
    norm_data_str = data_str.split(';')
    return norm_data_str[1]


key['date'] = key['16.02.2024;16.00'].apply(key_date)
key['value'] = key['16.02.2024;16.00'].apply(key_value)
key = key.drop('16.02.2024;16.00', axis = 1)
key.head()
```

```python
key.shape
```

```python
key['month'] = key['date'].dt.month
key['year'] = key['date'].dt.year
key.head()
```

```python
new_key = key.groupby(['year', 'month']).min('date')
new_key.shape
```

```python
import datetime

def key_date(date):
    date = str(date)
    date = date.split('.')
    new_date = datetime.date(int(date[1]),int(date[0]),1)
    return new_date


rate_rus['date'] = rate_rus['date'].apply(key_date)
rate_rus.head()
```
